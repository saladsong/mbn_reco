```python
from collections import Counter

def _get_fin_vocab(vocab):
    fin_vocab = {}
    
    c_vocab = Counter(vocab)
    cnt = len(fin_vocab)
    for voca in c_vocab:
        if (voca != '.') and (c_vocab[voca] > 1):
            fin_vocab[voca] = cnt
            cnt += 1
            
    return fin_vocab
```

```python
def _tok_to_idx(tok_plylst, fin_vocab):
    res_idx = []
    for tok in tok_plylst:
        if tok in fin_vocab:
            res_idx.append(fin_vocab[tok])
        else:
            pass
        
    return res_idx
```

```python
##

fin_vocab = _get_fin_vocab(vocab)
print('fin voca #:', len(fin_vocab))

txt_to_idx = []

for article in txt_to_tok:
    res_idx = _tok_to_idx(article, fin_vocab)
    txt_to_idx.append(res_idx)
```

```python
## build article - voca (sparse) matrix (as the source of tf-idf matrix)

def _preproc_for_csr(txt_to_idx, from_idx=0):
    user_lst = []
    vocab_lst = []

    for idx, title in enumerate(txt_to_idx[from_idx:]):
        uid = from_idx + idx
        uid_lst = []
        for tok in title:
            vocab_lst.append(tok)
            user_lst.append(uid)             

    return user_lst, vocab_lst
```

```python
user_lst, vocab_lst = _preproc_for_csr(txt_to_idx)

rows = np.array(user_lst)
cols = np.array(vocab_lst)
data = np.ones(len(user_lst))
data.shape
```

```python
import scipy.sparse as sp

usr_itm_mat = sp.csr_matrix((data, (rows, cols)), shape=(len(txt_to_tok), len(fin_vocab)))
usr_itm_mat.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

def _build_tfidf_mat(usr_itm_mat):
    tfidf_trans = TfidfTransformer()
    tfidf_mat = tfidf_trans.fit_transform(usr_itm_mat)
    
    return tfidf_mat
```

```python
tfidf_mat = _build_tfidf_mat(usr_itm_mat)
```

```python
## calculate similarity between articles

from sklearn.metrics.pairwise import linear_kernel

def _get_sim_article(tfidf_mat, given, topn):
    res_sim = dict()
    most_sim_idx, most_diff_idx = list(), list()

    cos_sim = linear_kernel(tfidf_mat[given:given+1], tfidf_mat).flatten()

    for idx, sim in enumerate(cos_sim):
        if (sim < 1) and (sim > 0):
            res_sim[sim] = idx

    res_sorted = sorted(res_sim, reverse=True)
    most_sim = res_sorted[:topn]
    most_diff = res_sorted[-topn:]

    for item in most_sim:
        most_sim_idx.append(res_sim[item])
        
    for item in most_diff:
        most_diff_idx.append(res_sim[item])
        
    return most_sim_idx, most_diff_idx
```
